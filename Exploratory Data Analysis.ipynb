#### Step 4: Analysis of Experience and Job Satisfaction

Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Clean YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
})

df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')


# 2. Create experience ranges

df['ExperienceRange'] = pd.cut(
    df['YearsCodePro'],
    bins=[0, 5, 10, 20, float('inf')],
    labels=['0-5 years', '5-10 years', '10-20 years', '20+ years'],
    right=False
)


# 3. Convert JobSat to numeric

df['JobSat_Score'] = pd.to_numeric(df['JobSat'], errors='coerce')


# 4. Keep only valid rows

plot_df = df.dropna(subset=['ExperienceRange', 'JobSat_Score']).copy()

print("Plot data shape:", plot_df.shape)

# 5. Summary table
summary = (
    plot_df.groupby('ExperienceRange', observed=False)['JobSat_Score']
    .agg(Median_JobSat='median', Count='count')
    .reset_index()
)

print(summary)


# 6. Plot median JobSat

plt.figure(figsize=(10, 6))

sns.barplot(
    data=summary,
    x='ExperienceRange',
    y='Median_JobSat',
    hue='ExperienceRange',
    legend=False,
    palette='Blues',
    order=['0-5 years', '5-10 years', '10-20 years', '20+ years']
)

plt.title('Median Job Satisfaction by Professional Experience', fontsize=16)
plt.xlabel('Years of Professional Coding', fontsize=12)
plt.ylabel('Median Job Satisfaction Score', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

#### Analyzing Remote Work Preferences by Job Role

Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Select only needed columns

# We keep only the columns needed for this task
df_remote = df[['RemoteWork', 'Employment']].copy()


# 2. Check missing values

# This shows how many missing values exist before cleaning
print("Missing values before cleaning:")
print(df_remote.isnull().sum())


# 3. Fill missing values

# Fill missing RemoteWork values with the most common category
df_remote['RemoteWork'] = df_remote['RemoteWork'].fillna(df_remote['RemoteWork'].mode()[0])

# Fill missing Employment values with 'Unknown'
df_remote['Employment'] = df_remote['Employment'].fillna('Unknown')


# 4. Split multiple employment types

# Some rows contain multiple values separated by semicolons
# Example: "Employed, full-time;Student, part-time"
df_remote['Employment'] = df_remote['Employment'].str.split(';')

# Explode turns each list into separate rows
df_remote = df_remote.explode('Employment')

# Reset index to avoid duplicate-label issues
df_remote = df_remote.reset_index(drop=True)

# Remove extra spaces from Employment values
df_remote['Employment'] = df_remote['Employment'].str.strip()

# 5. Count plot for RemoteWork

# This shows the overall distribution of remote work preferences
plt.figure(figsize=(10, 6))

sns.countplot(
    data=df_remote,
    x='RemoteWork',
    order=df_remote['RemoteWork'].value_counts().index
)

plt.title('Distribution of Remote Work Preferences')
plt.xlabel('Remote Work Type')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


# 6. Cross-tabulation (COUNTS)

# This is the actual crosstab table
crosstab_counts = pd.crosstab(
    df_remote['Employment'],
    df_remote['RemoteWork']
)

print("\nCross-tabulation (Counts):")
print(crosstab_counts)


# 7. Cross-tabulation (ROW PERCENTAGES)

# This shows percentages within each Employment group
crosstab_percent = pd.crosstab(
    df_remote['Employment'],
    df_remote['RemoteWork'],
    normalize='index'
) * 100

print("\nCross-tabulation (Percentages):")
print(crosstab_percent.round(2))


# 8. Optional: Heatmap of crosstab

# This is another way to visualize the cross-tabulation
plt.figure(figsize=(10, 6))

sns.heatmap(crosstab_percent.round(1), annot=True, cmap='Blues', fmt='.1f')

plt.title('Remote Work Preference by Employment Type (%)')
plt.xlabel('Remote Work Type')
plt.ylabel('Employment Type')
plt.tight_layout()
plt.show()
```

#### Analyzing Programming Language Trends by Region

Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. SELECT RELEVANT COLUMNS


# Keep only Country and Language column
df_lang = df[['Country', 'LanguageHaveWorkedWith']].copy()

# 2. HANDLE MISSING VALUES

# Drop rows where language or country is missing
df_lang = df_lang.dropna(subset=['Country', 'LanguageHaveWorkedWith'])

# 3. SPLIT MULTIPLE LANGUAGES

# Languages are separated by ';'
# Convert string → list
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')

# Explode so each language gets its own row
df_lang = df_lang.explode('LanguageHaveWorkedWith')

# Clean whitespace
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.strip()

# 4. SELECT REGIONS / COUNTRIES

# Example: choose top 5 countries by responses
top_countries = df_lang['Country'].value_counts().head(5).index

# Filter dataset to only those countries
df_filtered = df_lang[df_lang['Country'].isin(top_countries)]

# 5. COUNT TOP LANGUAGES PER COUNTRY

# Count occurrences of each language per country
lang_counts = (
    df_filtered
    .groupby(['Country', 'LanguageHaveWorkedWith'])
    .size()
    .reset_index(name='Count')
)

# 6. GET TOP 5 LANGUAGES PER COUNTRY

# Sort values by count
lang_counts = lang_counts.sort_values(['Country', 'Count'], ascending=[True, False])

# Keep top 5 languages per country
top_langs = lang_counts.groupby('Country').head(5)

print(top_langs.head(20))

# 7. CREATE HEATMAP DATA

# Pivot table: rows = Country, columns = Language
heatmap_data = pd.pivot_table(
    df_filtered,
    index='Country',
    columns='LanguageHaveWorkedWith',
    aggfunc='size',
    fill_value=0
)

# Keep only top 10 languages overall
top_languages = heatmap_data.sum().sort_values(ascending=False).head(10).index

heatmap_data = heatmap_data[top_languages]

# 8. PLOT HEATMAP

plt.figure(figsize=(12, 6))

sns.heatmap(
    heatmap_data,
    cmap='Blues',
    annot=True,
    fmt='d'
)

plt.title('Programming Language Popularity by Country', fontsize=16)
plt.xlabel('Programming Language')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

#### Educational Background and Employment Type

Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --------------------------------------------------
# 1. SELECT COLUMNS
# --------------------------------------------------
df_edu = df[['EdLevel', 'Employment']].copy()

# --------------------------------------------------
# 2. HANDLE MISSING VALUES
# --------------------------------------------------
df_edu['EdLevel'] = df_edu['EdLevel'].fillna('Unknown')
df_edu['Employment'] = df_edu['Employment'].fillna('Unknown')

# --------------------------------------------------
# 3. SIMPLIFY EDUCATION LEVELS
# --------------------------------------------------
def simplify_education(x):
    x = x.lower()

    if 'bachelor' in x:
        return 'Bachelor'
    elif 'master' in x:
        return 'Master'
    elif 'professional degree' in x or 'doctor' in x:
        return 'Postgrad'
    elif 'secondary' in x or 'primary' in x:
        return 'School'
    else:
        return 'Other'

df_edu['EdLevel_Short'] = df_edu['EdLevel'].apply(simplify_education)

# --------------------------------------------------
# 4. SIMPLIFY EMPLOYMENT TYPES
# --------------------------------------------------
# Split multi-values
df_edu['Employment'] = df_edu['Employment'].str.split(';')
df_edu = df_edu.explode('Employment')

# Clean spaces
df_edu['Employment'] = df_edu['Employment'].str.strip().str.lower()

def simplify_employment(x):
    if 'full-time' in x:
        return 'Full-time'
    elif 'part-time' in x:
        return 'Part-time'
    elif 'student' in x:
        return 'Student'
    elif 'not employed' in x:
        return 'Unemployed'
    elif 'independent contractor' in x or 'self-employed' in x:
        return 'Self-employed'
    else:
        return 'Other'

df_edu['Employment_Short'] = df_edu['Employment'].apply(simplify_employment)

# Reset index after explode
df_edu = df_edu.reset_index(drop=True)

# --------------------------------------------------
# 5. CROSS-TAB (PERCENTAGE)
# --------------------------------------------------
crosstab = pd.crosstab(
    df_edu['EdLevel_Short'],
    df_edu['Employment_Short'],
    normalize='index'
) * 100

print(crosstab.round(2))

# --------------------------------------------------
# 6. STACKED BAR (CLEAN VISUAL)
# --------------------------------------------------
crosstab.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6),
    colormap='Blues',
)

plt.title('Employment Distribution by Education Level (Simplified)')
plt.xlabel('Education Level')
plt.ylabel('Percentage (%)')

plt.xticks(rotation=0)
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1))
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# --------------------------------------------------
# 7. HEATMAP (VERY CLEAN)
# --------------------------------------------------
plt.figure(figsize=(8, 5))

sns.heatmap(
    crosstab.round(1),
    annot=True,
    cmap='Blues',
    fmt='.1f'
)

plt.title('Education vs Employment (%)')
plt.xlabel('Employment Type')
plt.ylabel('Education Level')

plt.tight_layout()
plt.show()
```
